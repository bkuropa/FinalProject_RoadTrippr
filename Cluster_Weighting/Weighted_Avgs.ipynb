```python
# Create a new weighted cluster for future use
wt_cluster_avg = clus_avg
```

```python
# Boosting cluster 1
wt_cluster_avg["Cluster1"][['pubs-bars','restaurants']] = clus_avg["Cluster1"][['pubs-bars','restaurants']]*1.06
print(wt_cluster_avg["Cluster1"][['pubs-bars','restaurants']])
wt_cluster_avg["Cluster1"][['malls','zoo']] = clus_avg["Cluster1"][['malls','zoo']]*1.06
print(wt_cluster_avg["Cluster1"][['malls','zoo']])
wt_cluster_avg["Cluster1"][['theatres','dance-clubs']] = clus_avg["Cluster1"][['theatres','dance-clubs']]*1.7
print(wt_cluster_avg["Cluster1"][['theatres','dance-clubs']])
```

```python
# Boosting cluster 2
wt_cluster_avg["Cluster2"][['theatres']] = clus_avg["Cluster2"][['theatres']]*1.125
print(wt_cluster_avg["Cluster2"][['theatres']])
```

```python
# Boosting cluster 3
wt_cluster_avg["Cluster3"][['cafes','bakeries']] = clus_avg["Cluster3"][['cafes','bakeries']]*2.1
print(wt_cluster_avg["Cluster3"][['cafes','bakeries']])
wt_cluster_avg["Cluster3"][['view-points','churches','monuments','gardens']] = clus_avg["Cluster3"][['view-points','churches','monuments','gardens']]*1.4
print(wt_cluster_avg["Cluster3"][['view-points','churches','monuments','gardens']])
```

```python
# Boosting cluster 4
wt_cluster_avg["Cluster4"]['art-galleries'] = clus_avg["Cluster4"]['art-galleries']*1.2
print(wt_cluster_avg["Cluster4"]['art-galleries'])
```
